```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
dataset=pd.read_csv('Pre_processdata.csv')
```

```python
dataset1=dataset.drop('date',1)
```

```python
dataset1
```

```python
datas=pd.get_dummies(dataset1,drop_first=True)
```

```python
datas
```

```python
datas.columns
```

```python
ind=datas.drop('Usage_kWh',1)
dep=datas['Usage_kWh'].astype(int)
```

```python
ind
```

```python
def select_k(ind,dep,n):
    selection=SelectKBest(score_func=chi2,k=n)
    fit_select=selection.fit(ind,dep)
    select_feature=fit_select.transform(ind)
    return select_feature
```

```python
def split_scaler(ind,dep):
    X_train, X_test, y_train, y_test = train_test_split(ind, dep, test_size = 0.25, random_state = 0)
    sc=StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test) 
    return X_train, X_test, y_train, y_test
```

```python
def r2score(regressor,X_test,y_test):
    y_pred=regressor.predict(X_test)
    from sklearn.metrics import r2_score
    r2=r2_score(y_test,y_pred)
    return r2
```

```python
def linear(X_train,y_train,X_test):       
        from sklearn.linear_model import LinearRegression
        regressor = LinearRegression()
        regressor.fit(X_train, y_train)
        r2=r2score(regressor,X_test,y_test)
        return  r2 
```

```python
def random(X_train,y_train,X_test):
    from sklearn.ensemble import RandomForestRegressor
    RandomForestRegressor(n_estimators = 10, random_state = 0)
    regressor.fit(X_train,y_train)
    r2=r2score(regressor,X_test,y_test)
    return r2
```

```python
def decision(X_train,y_train,X_test):
    from sklearn.tree import DecisionTreeRegressor
    regressor = DecisionTreeRegressor(random_state = 0)
    regressor.fit(X_train,y_train)
    r2=r2score(regressor,X_test,y_test)
    return r2
```

```python
def tables(r2lin,r2ran,r2deci):
    table=pd.DataFrame(index=['chi2_square'],columns=['r2lin','r2ran','r2deci'])
    for number,i in enumerate(table.index):
        table['r2lin'][i]=r2lin[number]
        table['r2ran'][i]=r2ran[number]
        table['r2deci'][i]=r2deci[number]
    return table
```

```python
kbest=select_k(ind,dep,7)
```

```python
X_train,X_test,y_train,y_test=split_scaler(kbest,dep)
```

```python
r2lin=[]
r2ran=[]
r2deci=[]
```

```python
X_train,X_test,y_train,y_test=split_scaler(kbest,dep)
r2_lin=linear(X_train,y_train,X_test)
r2lin.append(r2_lin)
r2_ran=linear(X_train,y_train,X_test)
r2ran.append(r2_ran)
r2_deci=linear(X_train,y_train,X_test)
r2deci.append(r2_deci)
   
```

```python
result=tables(r2lin,r2ran,r2deci)
```
