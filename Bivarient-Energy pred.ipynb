# remove multicolinearity using vif(variance inflation factor)

```python
def qualquan(dataset):
    qual=[]
    quan=[]
    for columnName in dataset.columns:
        if dataset[columnName].dtype=='O':
            qual.append(columnName)
        else:
            quan.append(columnName)
    return qual,quan
```

```python
qual,quan=qualquan(dataset)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor 
```

```python
def vif_fun(x):
    vif=pd.DataFrame()
    vif['variable']=x.columns
    vif['VIF']=[variance_inflation_factor(x.values,i)for i in range(x.shape[1])]
               
    return(vif)
```

```python
vif_fun(dataset[[
 'Lagging_Current_Reactive.Power_kVarh',
 'Leading_Current_Reactive_Power_kVarh',
 'CO2(tCO2)',
 
 'Leading_Current_Power_Factor',
 'NSM']])
```

# T-Test

# Independent sample - unpaired T-TEST
# different group same condition

```python
from scipy.stats import ttest_ind
light=dataset[dataset['Load_Type']=='Light_Load']['Usage_kWh']
medium=dataset[dataset['Load_Type']=='Medium_Load']['Usage_kWh']
ttest_ind(light,medium)
```

# depdent sample - paired T-Test
# same group different condition

```python
from scipy.stats import ttest_ind
light1=dataset[dataset['Load_Type']=='Light_Load']['Lagging_Current_Power_Factor']
light2=dataset[dataset['Load_Type']=='Light_Load']['Leading_Current_Power_Factor']
ttest_ind(light1,light2)
```
